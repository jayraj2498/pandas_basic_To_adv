## dropna  & fillna

```python
import pandas as pd
import numpy as np

# Create a DataFrame with random values and some null values
data = {
    'col1': np.random.randint(1, 100, 20),
    'col2': np.random.choice([1, 2, np.nan], 20),
    'col3': np.random.choice([5, 6, np.nan], 20),
    'col4': np.random.randint(100, 200, 20),
    'col5': np.random.choice([10, np.nan, 12], 20),
}

df_with_nulls = pd.DataFrame(data)

# Display the DataFrame
print(df_with_nulls)
```

```python
df = df_with_nulls
df
```

```python
# drop null value

df.dropna(axis=0)
```

```python
df.dropna(how='any')   # it will drop null of row 
```

```python
df.dropna(how='all') 
```

```python
df.dropna(axis=1)  # it drop along col 
```

```python
# along a column if you want to remove null value 
```

```python
df
```

```python
df.dropna(subset=['col3'])  # in col3 whichever val is nul , is get remove 
```

```python
# inplace it will remove null value and give you a last datafame which is unchangeble 
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df
```

```python
df.dropna(thresh=1)  # in particular row having singl nulll val that will be removed 
```

```python
df.dropna(thresh=2) 
```
