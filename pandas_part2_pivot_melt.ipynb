```python
import pandas as pd 
import numpy as np 
```

## melt

```python
var= pd.DataFrame({'days':[1,2,3,4,5,6] , 'eng':[11,12,13,14,15,16] , 'maths':[17,23,12,45,12,23]}) 

var
```

```python
# vartically reshape 

pd.melt(var)
```

```python
pd.melt(var,id_vars=['days'] , var_name='subject' , value_name='marks')
```

### pivot()

```python
var= pd.DataFrame({'days':[1,2,3,4,5,6] ,
                   'st_nmae': ['a','b','c','a','b','c'],
                   'eng':[11,12,13,14,15,16] ,
                   'maths':[17,23,12,45,12,23]}) 

var
```

```python
var.pivot(index='days'  , columns='st_nmae')
```

```python
var.pivot(index='days'  , columns='st_nmae', values='eng')
```

```python
var.pivot(index='days'  , columns='st_nmae', values='maths')
```

```python
var.pivot_table(index='st_nmae'  , columns='days', aggfunc='mean')
```

```python
var.pivot_table(index='st_nmae'  , columns='days', aggfunc='sum')
```
